```python


# Fill null values
for column in train_data.columns:
    if train_data[column].dtype == 'object':  # Categorical column
        train_data[column].fillna(train_data[column].mode()[0], inplace=True)
    else:  # Numerical column 
        train_data[column].fillna(train_data[column].mean(), inplace=True)
```

Handling the categorical values


```python
for i in train_data.columns:
    print(f"Number of unique values in '{i}':", train_data[i].nunique())
```

```python
train_data.drop(columns=['ID','efs_time'],inplace=True)
```

```python
categories = []
numerical = []

# Loop through all columns in the DataFrame
for col in train_data.columns:
    if train_data[col].dtype == 'object' or train_data[col].dtype.name == 'category':
        # If the column is categorical
        categories.append(col)
    else:
        # If the column is numerical
        numerical.append(col)

# Print the results
print("Categorical columns:", categories)
print("Number of categorical",len(categories))
print("Numerical columns:", numerical)
print("Number of numerical",len(numerical))
```

Convert categories to string

```python
for col in train_data.columns:
    if train_data[col].dtype == 'object' or train_data[col].dtype.name == 'category':
        # If the column is categorical, convert it to string directly in the DataFrame
        train_data[col] = train_data[col].astype(str)
```

Furthur preprocessing

```python
hla_columns = [
    'hla_high_res_8', 'hla_high_res_6', 'hla_high_res_10',
    'hla_low_res_6', 'hla_low_res_10', 'hla_low_res_8',
    'hla_match_c_high', 'hla_match_c_low', 'hla_match_a_high', 
    'hla_match_b_low', 'hla_match_a_low', 'hla_match_b_high'
]
```

```python
train_data['hla_avg_match'] = train_data[hla_columns].mean(axis=1)

# Example 2: Calculate the difference between high-res and low-res for each type of HLA match
train_data['hla_high_low_diff_6'] = train_data['hla_high_res_6'] - train_data['hla_low_res_6']
train_data['hla_high_low_diff_8'] = train_data['hla_high_res_8'] - train_data['hla_low_res_8']
train_data['hla_high_low_diff_10'] = train_data['hla_high_res_10'] - train_data['hla_low_res_10']

# Example 3: Binary features indicating whether the match is high (1) or low (0)
train_data['hla_match_c_high_flag'] = (train_data['hla_match_c_high'] > 0).astype(int)
train_data['hla_match_c_low_flag'] = (train_data['hla_match_c_low'] > 0).astype(int)
train_data['hla_match_a_high_flag'] = (train_data['hla_match_a_high'] > 0).astype(int)
train_data['hla_match_b_low_flag'] = (train_data['hla_match_b_low'] > 0).astype(int)
```

```python
train_numerical = []

for col in train_data.columns:
    if train_data[col].dtype == 'object' or train_data[col].dtype.name == 'category':
        continue
    else:
        train_numerical.append(col)
```
